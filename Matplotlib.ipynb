```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(2,9,10)
x
```

```python
y = x**2
y
```

```python
plt.plot(x,y, 'skyblue')
plt.title("My first plot")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,"g")

plt.subplot(1,2,2)
plt.plot(x,y,"red")
```

#### **Object Oriented Programming**

```python
figure = plt.figure()
plot = figure.add_axes([0.1,0.1,1,1])

plot.plot(x,y,"red")
```

```python
figure = plt.figure()

plot01 = figure.add_axes([0.1,0.1,0.9,0.9])
plot01.plot(x,y,"red")

plot02 = figure.add_axes([0.2,0.4,0.3,0.3])
plot02.plot(x,y,"blue")
```

### **Multiple plots**

```python
figure, plots = plt.subplots(nrows=2, ncols=2)

plots[0][0].plot(x,y,"b")
plots[0][1].plot(x,y,"r")
plots[1][0].plot(x,y,"g")
plots[1][1].plot(x,y,"y")

plots[0][0].set_title("Blue")
plots[0][1].set_title("Red")
plots[1][0].set_title("Green")
plots[1][1].set_title("Yellow")
```

### **Plot Size**

```python
figure = plt.figure(figsize=(8,4))

plot01 = figure.add_axes([0.1,0.1,0.9,0.9])
plot01.plot(x,y,"red")
```

### **2 Plots in 1 figure**

```python
figure = plt.figure()
plots = figure.add_axes([0.1,0.1,0.9,0.9])

plots.plot(x, x**2, label="Plot 01")
plots.plot(x, x**3, label="Plot 02")
plots.legend(loc=0)
```

### **Plot Colors**

```python
figure = plt.figure()
plots = figure.add_axes([0.1,0.1,0.9,0.9])
plots.plot(x, y, label="Plot 01", color="yellowgreen") # salmon, tomato, sienna, lime...
```

### **Lines**

```python
figure = plt.figure()
plots = figure.add_axes([0.1,0.1,0.9,0.9])
plots.plot(x, y, label="Plot 01", color="pink", linewidth=5, alpha=0.8, linestyle=":") # ":",  "-."
```

### **Markers**

```python
figure = plt.figure()
plots = figure.add_axes([0.1,0.1,0.9,0.9])
plots.plot(x, y, label="Plot 01", color="pink", linewidth=3, alpha=0.8, marker="o", markersize=5)
```
