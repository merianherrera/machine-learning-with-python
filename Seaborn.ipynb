```python
import seaborn as sns
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset("tips")
tips.head(10)
```

## Distribution Plots

```python
sns.distplot(tips["total_bill"])
```

```python
sns.distplot(tips["total_bill"], bins=20)
```

```python
sns.distplot(tips["tip"], bins=10)
```

```python
sns.jointplot(x="total_bill", y="tip", data=tips)
```

```python
sns.jointplot(x="total_bill", y="tip", data=tips, kind="hex")
```

```python
sns.jointplot(x="total_bill", y="tip", data=tips, kind="reg") 
# This one looks similar to scatter plot but this one shows the linear regression.
```

```python
sns.jointplot(x="total_bill", y="tip", data=tips, kind="kde") 
# This is a density plot, darker color, more data concentrated.
```

```python
sns.pairplot(tips)
```

```python
sns.pairplot(tips, hue="sex") # 'hue' is used to indicate the variable we want to separate data.
```

```python
sns.pairplot(tips, hue="day")
```

```python
sns.rugplot(tips["total_bill"])
```

```python
sns.rugplot(tips["size"])
```

## Heat Maps

```python
flights = sns.load_dataset("flights")
flights.head(10)
```

```python
flights_matrix = flights.pivot_table(index="month", columns="year", values="passengers")
flights_matrix
```

```python
sns.heatmap(flights_matrix)
```

```python
sns.heatmap(flights_matrix, cmap="viridis") # inferno, magma, coolwarm...
```

```python
sns.heatmap(flights_matrix, cmap="viridis", linecolor="white", linewidth="1")
```

## PairGrid and Facetgrid

```python
flowers = sns.load_dataset("iris")
flowers.head(10)
```

```python
sns.pairplot(flowers)
```

```python
plots = sns.PairGrid(flowers)
plots.map(plt.scatter)
```

```python
plots = sns.PairGrid(flowers)
plots.map_diag(sns.distplot)
plots.map_upper(plt.scatter)
plots.map_lower(sns.kdeplot)
```
